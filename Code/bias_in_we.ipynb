## **Finding most biased words (Gender)**
(using projection on gender subspace {man-woman})

```
vocabp = []
vocabm = []

professions = []
misc_words = []
with open('/content/drive/My Drive/datasets/professions.txt' ,'r') as fl:
    for word in fl:
        professions.append(word[:-1])    

with open('/content/drive/My Drive/datasets/misc_words.txt' ,'r') as fl:
    for word in fl:
        misc_words.append(word[:-1])

for word in professions:
    vocabp.append((word,model1[word]/npy.linalg.norm(model1[word])))
    
for word in misc_words:
    vocabm.append((word,model1[word]/npy.linalg.norm(model1[word])))
    
    
projectionp = []
projectionm = []


man = model1['man'] / npy.linalg.norm(model1['man'])
woman = model1['woman'] / npy.linalg.norm(model1['woman'])

gender_subspc = (man - woman)

for word in vocabp:
    pro = npy.dot(model1[word[0]] / npy.linalg.norm(model1[word[0]]),gender_subspc)
    projectionp.append((word[0],pro))
    
for word in vocabm:
    pro = npy.dot(model1[word[0]] / npy.linalg.norm(model1[word[0]]),gender_subspc)
    projectionm.append((word[0],pro))
    
    
biasedp = sorted(projectionp, reverse = True, key = lambda x : x[1])
biasedm = sorted(projectionm, reverse = True, key = lambda x : x[1])




print("\nBias in Professions in Model - decending order (top 10) (man-woman)\n")
ten_bias = biasedp[0:10]
for bias in ten_bias:
    print(bias)
    
print("\nBias in Misc. words in Model - descending order (top 10) (man-woman)\n")
ten_bias = biasedm[0:10]
for bias in ten_bias:
    print(bias)
```

## **Debiasing the model**

```
vocabp = []
vocabm = []

professions = []
misc_words = []
with open('/content/drive/My Drive/datasets/professions.txt' ,'r') as fl:
    for word in fl:
        professions.append(word[:-1])    

with open('/content/drive/My Drive/datasets/misc_words.txt' ,'r') as fl:
    for word in fl:
        misc_words.append(word[:-1])

for word in professions:
    vocabp.append((word,model1[word]/npy.linalg.norm(model1[word])))
    
for word in misc_words:
    vocabm.append((word,model1[word]/npy.linalg.norm(model1[word])))
    
    
projectionp_gender = []
projectionp_religion = []
projectionp_race = []

projectionm_gender = []
projectionm_religion = []
projectionm_race = []

hard_debiasp_gender = dict()
hard_debiasp_religion = dict()
hard_debiasp_race = dict()

hard_debiasm_gender = dict()
hard_debiasm_religion = dict()
hard_debiasm_race = dict()

debiased_professions_gender = []
debiased_professions_religion = []
debiased_professions_race = []

debiased_misc_words_gender = []
debiased_misc_words_religion = []
debiased_misc_words_race = []


man = model1['man'] / npy.linalg.norm(model1['man'])
woman = model1['woman'] / npy.linalg.norm(model1['woman'])
indian = model1['indian'] / npy.linalg.norm(model1['indian'])
american = model1['american'] / npy.linalg.norm(model1['american'])
muslim = model1['muslim'] / npy.linalg.norm(model1['muslim'])
christian = model1['christian'] / npy.linalg.norm(model1['christian'])


gender_subspc = (man - woman)
race_subspc = (indian - american)
religion_subspc = (muslim - christian)

bias_subspc = npy.dot(gender_subspc,race_subspc)
bias_subspc = npy.dot(bias_subspc,religion_subspc)


for word in vocabp:
    pro_biased = ( npy.dot(model1[word[0]],gender_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionp_gender.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_professions_gender.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,gender_subspc)
    hard_debiasp_gender[word[0]] = pro_debiased
    
    pro_biased = ( npy.dot(model1[word[0]],race_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionp_race.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_professions_race.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,race_subspc)
    hard_debiasp_race[word[0]] = pro_debiased

    pro_biased = ( npy.dot(model1[word[0]],religion_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionp_religion.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_professions_religion.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,religion_subspc)
    hard_debiasp_religion[word[0]] = pro_debiased

for word in vocabm:
    pro_biased = ( npy.dot(model1[word[0]],gender_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionm_gender.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_misc_words_gender.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,gender_subspc)
    hard_debiasm_gender[word[0]] = pro_debiased
    
    pro_biased = ( npy.dot(model1[word[0]],race_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionm_race.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_misc_words_race.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,race_subspc)
    hard_debiasm_race[word[0]] = pro_debiased

    pro_biased = ( npy.dot(model1[word[0]],religion_subspc) / npy.linalg.norm(model1[word[0]]) )
    projectionm_religion.append((word[0],pro_biased))
    pro_debiased = (model1[word[0]] - pro_biased) / npy.linalg.norm(model1[word[0]] - pro_biased)
    w_mean = npy.mean(pro_debiased)
    w_b_mean = npy.mean(pro_biased)
    nu = w_mean - w_b_mean
    pro_debiased = nu + npy.dot(npy.sqrt(1-npy.square(npy.linalg.norm(nu))) , ((pro_debiased - w_b_mean)/npy.linalg.norm(pro_debiased - w_b_mean)))
    debiased_misc_words_religion.append((word[0],pro_debiased))
    pro_debiased = npy.dot(pro_debiased,religion_subspc)
    hard_debiasm_religion[word[0]] = pro_debiased
    

bpg = sorted(projectionp_gender, reverse = True, key = lambda x : x[1])
biasedp_gender = []
debiasedp_gender = []
for x in bpg:
    if(x[1] > hard_debiasp_gender[x[0]]):
        biasedp_gender.append(x)
        debiasedp_gender.append((x[0],hard_debiasp_gender[x[0]]))
bmg = sorted(projectionm_gender, reverse = True, key = lambda x : x[1])
biasedm_gender = []
debiasedm_gender = []
for x in bmg:
    if(x[1] > hard_debiasm_gender[x[0]]):
        biasedm_gender.append(x)
        debiasedm_gender.append((x[0],hard_debiasm_gender[x[0]]))

bpr = sorted(projectionp_race, reverse = True, key = lambda x : x[1])
biasedp_race = []
debiasedp_race = []
for x in bpr:
    if(x[1] > hard_debiasp_race[x[0]]):
        biasedp_race.append(x)
        debiasedp_race.append((x[0],hard_debiasp_race[x[0]]))
bmr = sorted(projectionm_race, reverse = True, key = lambda x : x[1])
biasedm_race = []
debiasedm_race = []
for x in bmr:
    if(x[1] > hard_debiasm_race[x[0]]):
        biasedm_race.append(x)
        debiasedm_race.append((x[0],hard_debiasm_race[x[0]]))

bpr = sorted(projectionp_religion, reverse = True, key = lambda x : x[1])
biasedp_religion = []
debiasedp_religion = []
for x in bpr:
    if(x[1] > hard_debiasp_religion[x[0]]):
        biasedp_religion.append(x)
        debiasedp_religion.append((x[0],hard_debiasp_religion[x[0]]))
bmr = sorted(projectionm_religion, reverse = True, key = lambda x : x[1])
biasedm_religion = []
debiasedm_religion = []
for x in bmr:
    if(x[1] > hard_debiasm_religion[x[0]]):
        biasedm_religion.append(x)
        debiasedm_religion.append((x[0],hard_debiasm_religion[x[0]]))

print("\nBias in Professions in model based on Gender - decending order (top 10)\n")
ten_bias = biasedp_gender[0:10]
ten_debias = debiasedp_gender[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)

print("\nBias in Professions in model based on Race - decending order (top 10)\n")
ten_bias = biasedp_race[0:10]
ten_debias = debiasedp_race[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)

print("\nBias in Professions in model based on Religion - decending order (top 10)\n")
ten_bias = biasedp_religion[0:10]
ten_debias = debiasedp_religion[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)


print("\nBias in Misc. words in model based on Gender - decending order (top 10)\n")
ten_bias = biasedm_gender[0:10]
ten_debias = debiasedm_gender[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)

print("\nBias in Misc. words in model based on Race - decending order (top 10)\n")
ten_bias = biasedm_race[0:10]
ten_debias = debiasedm_race[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)

print("\nBias in Misc. words in model based on Religion - decending order (top 10)\n")
ten_bias = biasedm_religion[0:10]
ten_debias = debiasedm_religion[0:10]
for bias, debias in zip(ten_bias, ten_debias):
    print(bias,"==>",debias)
```
